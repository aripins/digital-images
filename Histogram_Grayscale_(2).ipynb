```python
import cv2
import matplotlib.pyplot as plt

# Fungsi untuk menampilkan histogram grayscale
def histogram_grayscale(image, title):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Mengonversi gambar menjadi grayscale
    plt.figure(figsize=(8, 6))
    plt.title(f"Histogram Grayscale - {title}")
    plt.xlabel("Intensitas Pixel")
    plt.ylabel("Jumlah Pixel")
    plt.hist(gray.ravel(), 256, [0, 256], color='gray', alpha=0.7)
    plt.grid(True)
    plt.show()

# Fungsi utama
def main():
    # Path gambar (masukkan path gambar yang akan digunakan)
    file_path = "ningning.jpeg"  # Ganti dengan path gambar Anda

    # Membaca gambar
    image = cv2.imread(file_path)

    if image is None:
        print("Gambar tidak ditemukan! Periksa kembali path gambar.")
        return

    # Menampilkan histogram grayscale untuk gambar
    histogram_grayscale(image, "Gambar")

if __name__ == "__main__":
    main()
```
